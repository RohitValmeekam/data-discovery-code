```python
import pandas as pd
#pip install pyxlsb
df_mw = pd.read_excel("2019 Commercial Building Inventory - Midwest.xlsb",engine='pyxlsb',sheet_name='City',nrows=100)
df_ne = pd.read_excel("2019 Commercial Building Inventory - Northeast.xlsb",engine='pyxlsb',sheet_name='City',nrows=100)
df_sa = pd.read_excel("2019 Commercial Building Inventory - South Atlantic.xlsb",engine='pyxlsb',sheet_name='City',nrows=100)
df_w = pd.read_excel("2019 Commercial Building Inventory - West.xlsb",engine='pyxlsb',sheet_name='City',nrows=100)
df_sc = pd.read_excel("2019 Commercial Building Inventory - South Central.xlsb",engine='pyxlsb',sheet_name='City',nrows=100)
df_mw.columns
```

```python
df_merged = pd.merge(df_mw, df_ne, on='city_name', how='outer', suffixes=('_mw', '_ne'))
df_merged = pd.merge(df_merged, df_sa, on='city_name', how='outer', suffixes=('_merged', '_sa'))
df_merged = pd.merge(df_merged, df_w, on='city_name', how='outer', suffixes=('_merged', '_w'))
df_merged = pd.merge(df_merged, df_sc, on='city_name', how='outer', suffixes=('_merged', '_sc'))

df_merged
```

```python
result_df = pd.concat([df_mw, df_ne, df_sa, df_w, df_sc], ignore_index=True)
result_df
```
